```python
import re
import pandas as pd 
from pprint import pprint

from nltk.corpus import stopwords
stopwords = stopwords.words('spanish')

import pyLDAvis.gensim_models
from gensim.models import LdaModel
from gensim.corpora import Dictionary

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
data = pd.read_csv("C:/Users/andre/downloads/reviews_vidjew_es.csv")
data.head()
```

###  `[12 pts]` Punto 1: Hacer pre-procesamiento del texto

```python
def pre_procesado(texto):
    texto = texto.lower()
    texto = re.sub(r"[\W\d]+", " ", texto)
    texto = [palabra for palabra in texto.split() if palabra not in stopwords]
    return texto 

data['P-P'] = data['review_body'].apply(lambda texto: pre_procesado(texto))

data.head()
```

###  `[10 pts]` Punto 2: Modelo de LDA

```python
diccionario = Dictionary(data['P-P'].values)
diccionario.filter_extremes(no_below=5, no_above=0.6)
corpus = [diccionario.doc2bow(text) for text in data['P-P'].values]
modelo = LdaModel(corpus = corpus, id2word = diccionario, num_topics = 7, passes=50)
```

###  `[25 pts]` Punto 3: Visualización de LDA

```python
modelo.print_topics(num_words=10)
lda_display = pyLDAvis.gensim_models.prepare(modelo, corpus, diccionario)
pyLDAvis.display(lda_display)
```
