```python
import re
import nltk
import pandas as pd
import numpy as np
from collections import Counter
import texthero as hero

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')

from sklearn.feature_extraction.text import CountVectorizer
```

# Punto 1: Pre-Procesamiento

- `[5 pts]` Leer el archivo `dialogos.csv` usando `pandas`
- `[10 pts]` Crear una nueva columna con el texto en minúscula, sin caracteres especiales ni números y sin palabras vacias
- `[7 pts]` Usando la nueva columna, ¿cuáles son las 10 palabras más usadas? 

```python
dialogos= pd.read_csv('C:/Users/andre/Downloads/Dialogos.csv')
dialogos
```

```python
dial=dialogos["Locución"].str.lower()
dial=dial.apply(lambda x: re.sub(r"[\W\d]", " ",x))
dial=dial.apply(lambda x:[palabra for palabra in x.split() if palabra not in stopwords_sp])
dialogos['preprocesado']=dial.apply(lambda x:str(x)[1:-1]) 
dialogos
```

```python
vectorizer = CountVectorizer()

matrix = vectorizer.fit_transform(dialogos.preprocesado.values)
counts = pd.DataFrame(matrix.toarray(),
                      columns=vectorizer.get_feature_names())
cuentas=counts.sum(axis=0)
type(cuentas)

cuentas.sort_values(ascending=False).head(10)
```
